```python
from collections import Counter
import numpy as np
```

```python
data['Item']=data['StockCode'].astype(str)+ ' - '+ data['Description']
```

```python
basket_data=data.groupby('InvoiceNo')['Item'].agg(list).reset_index()
```

```python
size_distribution.columns=['Basket_size', 'Count']
size_distribution['Percentage']=size_distribution['Count']/len(basket_size)*100
size_distribution.head(5)
```

```python
# frequencies of items
all_items=[item for basket in basket_data['Item'] for item in basket]
item_freq=Counter(all_items)
```

```python
item_freq
```

```python
# most common items
most_common=pd.DataFrame(item_freq.most_common(20), columns=['Item', 'Freq'])
most_common
```

```python
total_baskets=len(basket_data)
total_baskets
```

```python
item_supp={item: count/total_baskets for item, count in item_freq.items()}
item_supp
```

```python
# items with min support of 1%
min_supp=0.01
supp_items={item: support
            for item, support in item_supp.items()
            if support >= min_supp}
supported_items_data=pd.DataFrame(supp_items.items(), columns=['Item', 'Support']).sort_values('Support', ascending=False)
```

```python
supported_items_data.head(5)
```

```python
# items commonly purchased with most freq item
target_item=most_common['Item'].iloc[0]
target_basket=basket_data[basket_data['Item'].apply(lambda x: target_item in x)]

target_item
```

```python
# counting co-occurring items
co_occurring=Counter([item
                      for basket in target_basket['Item']
                      for item in basket
                      if item != target_item])
```

```python
# co-occurrence support
co_occur_data=pd.DataFrame(co_occurring.items(), columns=['Item', 'Co_occur'])
co_occur_data['Support']=(co_occur_data['Co_occur']/len(target_basket))
```

```python
co_occur_data=co_occur_data.sort_values('Co_occur', ascending=False)
```

```python
co_occur_data.head(5)   # top 5 items commonly purchased with
                        # "WHITE HANGING HEART T-LIGHT HOLDER"
                        # i.e. the target_item
```

```python
while True:
  print("Recommendation system: ")
  # for i, item in enumerate(available_items[:5], 1):
  #   print(f"{i}.{item}")
  # print("... and many more")

  print("\1. Enter name to search")
  print("2. View more items")
  print("3. Exit")

  choice=input("Enter your choice(1-3): ")
  if choice=='1':
    search_term=input("\nEnter item name or code: ").lower()
    matching_items=[item for item in available_items
                    if search_term in item.lower()]
    if not matching_items:
      print("\nNo results found!")
      print("\nPress enter to continue")
      continue

    print("\nMatching items: ")
    for i, item in enumerate(matching_items[:10], 1):
      print(f"{i}. {item}")

    if len(matching_items)>100:
      print(f"... and {len(matching_items)-100} more items")

    try:
      item_index=int(input("\nEnter the number of your item(0 to go back): "))-1
      if item_index==-1:
        continue
      selected_item=matching_items[item_index]


      target_baskets=basket_data[basket_data['Item'].apply(lambda x: selected_item in x)]
      if len(target_basket)==0:
        print("\nItem not found in dataset")
        input("\nPress enter to continue ")
        continue

      co_occurring=Counter([item
                            for basket in target_basket['Item']
                            for item in basket
                            if item!=selected_item])

      co_occur_data['Support']=(co_occur_data['Co_occur']/len(target_basket)*100).round(2)

      print("\nRecommended items: ")
      for index, (_, row) in enumerate(co_occur_data.sort_values('Support', ascending=False).head(5).iterrows(), 1):
        print(f"{index}. {row['Item']}")
        print(f"Support: {row['Support']}% (Found together in {row['Co_occur']} baskets)")

    except(ValueError, IndexError):
      print("\nInvalid selection!")
    input("\nPress enter to continue ")

  elif choice=='2':
    start_index=5
    for i, item in enumerate(available_items[start_index: start_index+10], start_index+1):
      print(f"{i}. {item}")
    input("\nPress enter to continue ")

  elif choice=='3':
    print("\nThank you for using recommendation system!")
    break

  else:
    print("\nInvalid choice! Please enter a number from 1-3")
    input("\nPress enter to continue")
```
